```python
import pandas as pd
from math import log
from sklearn.feature_extraction.text import TfidfVectorizer
from tqdm import tqdm
```

```python
df = pd.read_excel("news.xlsx")
df = df.fillna('')
title = df["title"].tolist()
content = df["content"].tolist()

docs = title + content
```

```python
corpus = docs

tfidfv = TfidfVectorizer(sublinear_tf=True).fit(corpus)
tfidfv_array = pd.DataFrame(tfidfv.transform(corpus).toarray())
tfidfv_voca = pd.DataFrame(sorted(tfidfv.vocabulary_.items(), reverse=True))

print(tfidfv_voca[0].tolist())

tfidfv_voca.to_csv("tf_idf.csv")
```
